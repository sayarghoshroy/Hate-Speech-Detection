```
import pickle
import json 
import requests
import time
```

```
f = open('2020_processed_data/ge.pickle', 'rb')
data = pickle.load(f)
```

```
full_tweet = data['full_tweet']
tweet_raw_text = data['tweet_raw_text']
task1 = data['task_1']
```

```
api_key = 'AIzaSyB-kA0qmu96oINfIE_2knvUxMyvnpo5FQ8'
url = ('https://commentanalyzer.googleapis.com/v1alpha1/comments:analyze' +    
    '?key=' + api_key)
lg = ['de']
attr = ['TOXICITY', 'SEVERE_TOXICITY', 'IDENTITY_ATTACK', 'INSULT', 'PROFANITY', 'THREAT',]
attr_dict = {}
attr_results = {}
for i in attr:
    attr_dict[i] = {}
    attr_results[i+'_WHOLE'] = []
    attr_results[i+'_RAW'] = []
```

```
count = 0
for i in range(len(full_tweet)):
    data_dict = {
        'comment': {'text': full_tweet[i]},
        'languages': lg,
        'requestedAttributes': attr_dict
    }
    count += 1
    time.sleep(2)
    response = requests.post(url=url, data=json.dumps(data_dict)) 
    response_dict = json.loads(response.content)
    for i in attr:
        attr_results[i+'_WHOLE'].append(response_dict["attributeScores"][i]["summaryScore"]["value"])
    if count % 100 == 0:
        print("Sentences processed:", count)
    # print(full_tweet[i], response_dict["attributeScores"]["TOXICITY"]["summaryScore"]["value"], task1[i])
    # print(json.dumps(response_dict, indent=2))
```

```
count = 0
for i in range(len(tweet_raw_text)):
    data_dict = {
        'comment': {'text': tweet_raw_text[i]},
        'languages': lg,
        'requestedAttributes': attr_dict
    }
    response = requests.post(url=url, data=json.dumps(data_dict)) 
    response_dict = json.loads(response.content)
    time.sleep(1.5)
    count += 1
    for i in attr:
        attr_results[i+'_RAW'].append(response_dict["attributeScores"][i]["summaryScore"]["value"])
    if count % 100 == 0:
        print("Sentences processed:", count)
    # print(full_tweet[i], response_dict["attributeScores"]["TOXICITY"]["summaryScore"]["value"], task1[i])
    # print(json.dumps(response_dict, indent=2))
```
